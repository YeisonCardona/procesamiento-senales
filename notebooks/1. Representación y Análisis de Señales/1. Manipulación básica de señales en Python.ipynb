## 1. Importación de Bibliotecas

Primero, importaremos las bibliotecas necesarias para trabajar con señales.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 2. Generación de Señales Básicas

### Señal Sinusoidal

Vamos a crear una señal sinusoidal utilizando `NumPy`. Definiremos la frecuencia, amplitud y fase de la señal.

```python
# Parámetros de la señal
frecuencia = 5        # Frecuencia en Hz
amplitud = 1          # Amplitud de la señal
fase = 0              # Fase en radianes
duracion = 1          # Duración en segundos
fs = 100              # Frecuencia de muestreo en Hz

# Creación del vector de tiempo
t = np.linspace(0, duracion, int(fs * duracion), endpoint=False)

# Generación de la señal sinusoidal
senal_seno = amplitud * np.sin(2 * np.pi * frecuencia * t + fase)

# Visualización
plt.plot(t, senal_seno)
plt.title("Señal Sinusoidal")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()
```

### Señal Cuadrada

También podemos generar una señal cuadrada utilizando `scipy.signal`.

```python
from scipy import signal

# Generación de la señal cuadrada
senal_cuadrada = signal.square(2 * np.pi * frecuencia * t)

# Visualización
plt.step(t, senal_cuadrada)
plt.title("Señal Cuadrada")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()
```

### Señal Triangular

La señal triangular es una forma de onda continua similar a la señal cuadrada, pero con transiciones lineales. Es útil en aplicaciones de modulación.

```python
# Generación de la señal triangular
senal_triangular = signal.sawtooth(2 * np.pi * frecuencia * t, 0.5)

# Visualización
plt.plot(t, senal_triangular)
plt.title("Señal Triangular")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()
```

### Señal de Onda Sierra

La señal de onda sierra es otra forma de onda periódica que se utiliza en aplicaciones de análisis de Fourier y generación de armónicos.


```python
# Generación de la señal de onda sierra
senal_sierra = signal.sawtooth(2 * np.pi * frecuencia * t)

# Visualización
plt.plot(t, senal_sierra)
plt.title("Señal de Onda Sierra")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()
```

### Señal de Ruido Gaussiano

El ruido gaussiano es una señal de ruido con una distribución normal. Se utiliza frecuentemente para simular perturbaciones en sistemas de comunicación.


```python
# Generación de ruido gaussiano
media = 0
desviacion_estandar = 0.5
senal_ruido_gaussiano = np.random.normal(media, desviacion_estandar, t.shape)

# Visualización
plt.plot(t, senal_ruido_gaussiano)
plt.title("Señal de Ruido Gaussiano")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()
```

### Señal de Ruido Uniforme

El ruido uniforme es un tipo de ruido aleatorio donde todos los valores dentro de un rango tienen la misma probabilidad de ocurrir. Es útil para simular entornos no gaussianos.


```python
# Generación de ruido uniforme
amplitud_min = -1
amplitud_max = 1
senal_ruido_uniforme = np.random.uniform(amplitud_min, amplitud_max, t.shape)

# Visualización
plt.plot(t, senal_ruido_uniforme)
plt.title("Señal de Ruido Uniforme")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()
```

## 3. Operaciones Básicas con Señales

### Suma de Señales

A continuación, sumaremos dos señales (una sinusoidal y una cuadrada) para observar el resultado.

```python
# Parámetros de las señales
frecuencia1 = 5
frecuencia2 = 10

# Señales sinusoidales de diferentes frecuencias
senal1 = np.sin(2 * np.pi * frecuencia1 * t)
senal2 = np.sin(2 * np.pi * frecuencia2 * t)

# Suma de señales
senal_sumada = senal1 + senal2

# Visualización
plt.plot(t, senal_sumada)
plt.title("Suma de Señales Sinusoidales")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()
```

### Multiplicación de Señales

Realicemos ahora la multiplicación de la señal sinusoidal con una versión escalada de sí misma.

```python
# Multiplicación de las dos señales sinusoidales
senal_multiplicada = senal1 * senal2

# Visualización
plt.plot(t, senal_multiplicada)
plt.title("Multiplicación de Señales Sinusoidales")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()
```

### Desplazamiento en el Tiempo

El desplazamiento en el tiempo de una señal permite mover la señal hacia adelante o hacia atrás en el tiempo.


```python
# Desplazamiento de la señal en el tiempo
desplazamiento = 0.05  # Desplazamiento en segundos
senal_desplazada = np.sin(2 * np.pi * frecuencia1 * (t - desplazamiento))

# Visualización
plt.plot(t, senal1, label="Original")
plt.plot(t, senal_desplazada, label="Desplazada")
plt.title("Desplazamiento en el Tiempo")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.legend()
plt.grid(True)
plt.show()
```

### Inversión en el Tiempo

La inversión en el tiempo refleja una señal a través del eje vertical, invirtiendo su orden temporal.


```python
# Inversión de la señal en el tiempo
senal_invertida = np.sin(2 * np.pi * frecuencia1 * -t)

# Visualización
plt.plot(t, senal1, label="Original")
plt.plot(t, senal_invertida, label="Invertida")
plt.title("Inversión en el Tiempo")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.legend()
plt.grid(True)
plt.show()
```

## 4. Transformación de Señales: Escalado en el Tiempo y Amplitud

Podemos modificar las señales escalando su amplitud o su duración.

### Escalado de Amplitud

```python
# Parámetro de escalado de amplitud
factor_amplitud = 2
senal_escalada_amplitud = factor_amplitud * senal1

# Visualización
plt.plot(t, senal1, label="Original")
plt.plot(t, senal_escalada_amplitud, label=f"Escalada (Factor {factor_amplitud})")
plt.title("Escalado de Amplitud")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.legend()
plt.grid(True)
plt.show()
```

### Escalado en el Tiempo

Para comprimir o expandir una señal en el tiempo, podemos ajustar la frecuencia de la señal.

```python
# Parámetro de escalado en el tiempo
factor_tiempo = 0.5
senal_escalada_tiempo = np.sin(2 * np.pi * frecuencia1 * t * factor_tiempo)

# Visualización
plt.plot(t, senal1, label="Original")
plt.plot(t, senal_escalada_tiempo, label=f"Escalada en Tiempo (Factor {factor_tiempo})")
plt.title("Escalado en el Tiempo")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.legend()
plt.grid(True)
plt.show()
```

### Compresión Temporal

La compresión temporal hace que la señal se “comprima” en el tiempo, lo cual aumenta su frecuencia aparente. Esto puede lograrse escalando el tiempo con un factor mayor que 1.

```python
# Factor de compresión temporal
factor_compresion = 2
senal_compresion = np.sin(2 * np.pi * frecuencia1 * t * factor_compresion)

# Visualización
plt.plot(t, senal1, label="Original")
plt.plot(t, senal_compresion, label=f"Compresión Temporal (Factor {factor_compresion})")
plt.title("Compresión Temporal")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.legend()
plt.grid(True)
plt.show()
```

### Expansión Temporal

La expansión temporal hace que la señal se “extienda” en el tiempo, disminuyendo su frecuencia aparente. Esto se logra escalando el tiempo con un factor menor que 1.


```python
# Factor de expansión temporal
factor_expansion = 0.5
senal_expansion = np.sin(2 * np.pi * frecuencia1 * t * factor_expansion)

# Visualización
plt.plot(t, senal1, label="Original")
plt.plot(t, senal_expansion, label=f"Expansión Temporal (Factor {factor_expansion})")
plt.title("Expansión Temporal")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.legend()
plt.grid(True)
plt.show()
```
