```python
from dsp_utils import SignalGenerator
from dsp_utils import Visualizer

from matplotlib import pyplot as plt
from ipywidgets import interact
import numpy as np
```

```python
f = [50]

@interact(n=(1, 100), phase=(0, np.pi, 0.1))
def _(n=1, phase=0):

    
    generador = SignalGenerator(frequency=f, amplitude=1, duration=1, sampling_rate=1000, phase=phase)
    vis = Visualizer(figsize=(16, 7), dpi=100)
    senal_seno = generador.sine_wave
    
    vis.plot_signal(senal_seno, time=generador.time, title="Señal Sinusoidal", alpha=0.5)
    plt.plot(generador.time[::n], senal_seno[:, ::n][0], marker='.', color='C1')
    plt.show()
```
