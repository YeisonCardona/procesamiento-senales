```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Dos variables normales correlacionadas
np.random.seed(42)
x = np.random.normal(0, 1, 1000)
y = 0.8 * x + np.random.normal(0, 0.5, 1000)

print("Correlación:", np.corrcoef(x, y)[0, 1])
```

```python
sns.displot([x, y])
```

```python
sns.jointplot([x, y])
```
