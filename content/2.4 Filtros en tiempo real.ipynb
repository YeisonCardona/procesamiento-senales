# Filtros en Tiempo Real

```python
from dsp_utils.signals import SignalGenerator
from dsp_utils.signals import Visualizer
```

```python
from scipy.signal import hilbert
import numpy as np
from matplotlib import  pyplot as plt
from ipywidgets import interact
```

```python
def control(x, y):
    @interact(t = (1, len(x)-1, 1))
    def _(t=1):
        plt.figure(figsize=(12, 8))
        plt.plot(x[:t], y[:t], '-')
        plt.xlim(x.min()-0.1, x.max()+0.1)
        plt.ylim(y.min()-0.1, y.max()+0.1)
        plt.plot([x[t], x[t-1]], [y[t], y[t-1]], linestyle='--', color='r', marker='o')
        plt.show()
```

```python
f = [5, [1, 25],]

generador = SignalGenerator(frequency=f, amplitude=1, duration=1, sampling_rate=1000)
vis = Visualizer(figsize=(16, 7), dpi=100)
senal_seno = generador.sine_wave
```

```python
i = 0
```

```python
plt.figure(figsize=(18, 8))
plt.plot(hilbert(senal_seno[i]).real)
plt.plot(hilbert(senal_seno[i]).imag)
```

```python
vis.plot_dsp(senal_seno[i], sample_rate=1000, time=generador.time, band=[-30, 30])
```

```python
vis.plot_dsp(hilbert(senal_seno[i]), sample_rate=1000, time=generador.time, band=[-30, 30])
```

```python
plt.figure(figsize=(18, 8))
plt.plot(hilbert(senal_seno[i]).real, hilbert(senal_seno[i]).imag, '-')
plt.axis('equal')
```

```python
plt.figure(figsize=(18, 8))
plt.plot(np.angle(hilbert(senal_seno[i])), '.')
```

```python
plt.figure(figsize=(18, 8))
plt.plot(np.unwrap(np.angle(hilbert(senal_seno[i]))))
```

```python
plt.figure(figsize=(18, 8))
plt.plot(np.diff(np.unwrap(np.angle(hilbert(senal_seno[i])))))
```
